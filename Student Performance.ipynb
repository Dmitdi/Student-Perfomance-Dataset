### 2. Описание датасета ###

Источник: Kaggle 
https://www.kaggle.com/datasets/whenamancodes/student-performance

**Student Performance.** Эти данные показывают успеваемость учащихся в средней школе двух португальских школ. Атрибуты данных включают оценки учащихся, демографические, социальные и школьные особенности, и они были собраны с использованием школьных отчетов и анкет. Предоставляется набора данных об успеваемости по математике (mat).

Условные обозначения исходного датасета (30 переменных)

1. **school** - школа («GP» - Габриэль Перейра/«MS» - Мусиньо да Силвейра)
2. **sex** - пол учащегося («Ж» - женский/«М» - мужской)
3. **age** - возраст студента (числовое: от 15 до 22)
4. **adress** - тип домашнего адреса учащегося («U» - городской/«R» - сельский)
5. **famsize** - размер семьи («LE3» — меньше или равно 3/«GT3» — больше 3)
6. **Pstatus** - статус совместного проживания родителей («Т» - совместное проживание/«А» - раздельно)
7. **Medu** - образование матери (числовое: 0 - нет, 1 - начальное образование (4 класс), 2 - 5-9 класс, 3 - среднее или 4 - высшее)
8. **Fedu** - образование отца (числовое: 0 - нет, 1 - начальное образование (4 класс), 2 - 5-9 класс, 3 - среднее или 4 - высшее)
9. **Mjob** - работа матери (номинальное: «учитель», «здравоохранение», гражданская «служба» (например, административная или полицейская), «домашний» или «другой»)
10. **Fjob** - работа отца (номинальное: «учитель», «здравоохранение», гражданская «служба» (например, административная или полицейская), «домашний» или «другой»)
11. **reason** - причина выбора этой школы (номинальное значение: близость к «дому», «репутация школы», предпочтение «курса» или «другое»)
12. **guardian** - опекун ученика (имя: «мать», «отец» или «другой»
13. **traveltime** - время в пути от дома до школы (числовое значение: 1 – <15 мин., 2 – от 15 до 30 мин., 3 – от 30 мин. до 1 часа или 4 – >1 часа)
14. **studytime** - еженедельное время обучения (числовое значение: 1 – <2 часов, 2 – от 2 до 5 часов, 3 – от 5 до 10 часов или 4 – >10 часов)
15. **failures** - количество прошлых сбоев класса (числовое: n, если 1<=n<3, иначе 4)
16. **schoolsup** - дополнительная образовательная поддержка (двоичный код: да или нет)
17. **famsup** - семейная образовательная поддержка (двоичный код: да или нет)
18. **paid** - дополнительные оплачиваемые занятия в рамках предмета курса (математика или португальский язык) (двоичный код: да или нет)
19. **activities** - внеклассные мероприятия (бинарные: да или нет)
20. **nurcery** - посещал детский сад (двоичный код: да или нет)
21. **higher** - хочет получить высшее образование (бинарное: да или нет)
22. **internet** - Доступ в Интернет дома (бинарный: да или нет)
23. **romantic** - с романтическими отношениями (бинарный: да или нет)
24. **famrel** - качество семейных отношений (числовое: от 1 - очень плохо до 5 - отлично)
25. **freetime** - свободное время после школы (числовое: от 1 - очень низко до 5 - очень высоко)
26. **goout** - гулять с друзьями (числовое значение: от 1 — очень низкий до 5 — очень высокий)
27. **Dalc** - потребление алкоголя в рабочие дни (числовое значение: от 1 — очень низкое до 5 — очень высокое)
28. **Walc** - потребление алкоголя в выходные дни (числовое значение: от 1 — очень низкое до 5 — очень высокое)
29. **health** - текущее состояние здоровья (числовое: от 1 - очень плохо до 5 - очень хорошо)
30. **absences** - количество пропусков занятий (числовое: от 0 до 93)

Наиболее показательными являются финальные оценки G3 – оставим на графиках только их.

```python
#импортируем все необходимые библиотеки
import pandas as pd
import numpy as np

#импортируем библиотеки для визуализации
import matplotlib.pyplot as plt
import seaborn as sns
```

### Готовим данные к работе ###

```python
#открываем файл
scores = pd.read_excel('~/Desktop/Python/Maths.xlsx')
scores.head(5)
```

```python
#ищем пропуски в данных
scores.isnull().sum()
```

```python
#ищем количество уникальных данных в каждом столбце
scores.nunique()
```

# Для наглядности строим графики, чтобы увидеть взаимосвязи между интересными нам перменными #

```python
bins = np.linspace(0, 20, 20)

plt.hist([scores[scores['sex'] == i]['G3'] for i in scores['sex'].unique()], 
             bins, alpha=0.5, density='True', label=scores['sex'].unique())
plt.legend(loc="upper right")
plt.ylabel('Доля', fontsize = 12)
plt.xlabel('Оценка', fontsize = 12)
plt.title('Распределение оценок в зависимости от пола', fontsize = 14)
plt.show()
```

Видим, что среди категории студентов с высокими оценками преобладают лица мужского пола.

```python
bins = np.linspace(0, 20, 20)

plt.hist([scores[scores['higher'] == i]['G3'] for i in scores['higher'].unique()], 
             bins, alpha=0.5, density='True', label=scores['higher'].unique())
plt.legend(loc="upper right")
plt.ylabel('Доля', fontsize = 12)
plt.xlabel('Оценка', fontsize = 12)
plt.title('Распределение оценок в зависимости от намерения продолжать учебу', fontsize = 14)
plt.show()
```

Видим, что среди категории студентов с высокими оценками преобладают лица, желающие продолжать обучение.

```python
bins = np.linspace(0, 20, 20)

plt.hist([scores[scores['internet'] == i]['G3'] for i in scores['internet'].unique()], 
             bins, alpha=0.5, density='True', label=scores['internet'].unique())
plt.legend(loc="upper right")
plt.ylabel('Доля', fontsize = 12)
plt.xlabel('Оценка', fontsize = 12)
plt.title('Распределение оценок в зависимости от наличия интернета', fontsize = 14)

plt.show()
```

Видим, что среди категории студентов с высокими оценками намного меньше людей, у которых нет интернета.

```python
bins = np.linspace(0, 20, 20)

plt.hist([scores[scores['romantic'] == i]['G3'] for i in scores['romantic'].unique()], 
             bins, alpha=0.5, density='True', label=scores['romantic'].unique())
plt.legend(loc="upper right")
plt.ylabel('Доля', fontsize = 12)
plt.xlabel('Оценка', fontsize = 12)
plt.title('Распределение оценок в зависимости от наличия отношений', fontsize = 14)
plt.show()
```

Видим, что среди категории студентов с высокими оценками очень мало тех, кто состоит в отношениях - в отличие от других групп.

```python
categorial_df = scores[['Medu', 'Fedu', 'Mjob', 'Fjob', 'Dalc', 'Walc']]
plt.figure(figsize = (20, 20))

for idx, col in enumerate(categorial_df):
    plt.subplot(5, 2, idx + 1)
    ax = sns.boxplot(data = categorial_df, y = scores['G3'], x = categorial_df[col], palette="Blues")
    plt.axhline(scores['G3'].median(), color='orange', linewidth=3) #Медиана
    plt.ylabel('Оценка', fontsize = 12)
    plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.2)
```

Видим, что уровень образования и работа родителей практически не влияют на распределение оценок среди студентов.

```python
categorial_df = scores[['traveltime', 'studytime', 'freetime', 'goout']]
plt.figure(figsize = (20, 20))

for idx, col in enumerate(categorial_df):
    plt.subplot(5, 2, idx + 1)
    ax = sns.boxplot(data = categorial_df, y = scores['G3'], x = categorial_df[col], palette="Blues")
    plt.axhline(scores['G3'].median(), color='orange', linewidth=3) #Медиана
    plt.ylabel('Оценка', fontsize = 12)
    plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.2)
```

В категориях "путь от дома до школы", "время, посвящаемое учебе", "количество свободного времени" и "прогулки" видим интересные зависимости: 
- чем больше времени студент тратит на дорогу до учебы, тем хуже оценки
- не так сильно влияет время, которое студент тратит на учебу на его оценки, хоть и небольшая корреляция есть
- чем больше у студента времени на прогулки, тем хуже его оценки
- свободное время не сильно коррелирует с оценкой студента

Из графика мы видим, что количество прогулов почти не влияет на распределение оценок.

Последний вопрос исследования - "влияет ли возраст на распределение оценок". Ответ: нет, среди разных возрастных категорий примерно одинаковое распределение оценок.
